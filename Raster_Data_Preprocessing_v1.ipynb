### This notebook is used to create a binary raster file by applying a threshold to FLT raster files generated from PCSWMM outputs.

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
```

```python
flt_path = "Data_Sample/thirteen.FLT"
```

```python
# Load the FLT raster
with rasterio.open(flt_path) as src:
    data = src.read(1)  # Read the band
    profile = src.profile  # Keep metadata 
    nodata = src.nodata

plt.imshow(data, cmap='Blues')
```

```python
data_valid = np.ma.masked_equal(data, nodata)

total_cells = data_valid.count()
cells = np.sum(data_valid > 2)

min_val = data_valid.min()
max_val = data_valid.max()
mean_val = data_valid.mean()

# Print results
print(f"Total valid cells: {total_cells}")
print(f"Cells with value > 0.1: {cells}")
print(f"Minimum value: {min_val}")
print(f"Maximum value: {max_val}")
print(f"Average value: {mean_val}")
```

```python
# Define a threshold
threshold = 1
binary_grid = (data > threshold).astype(np.uint8)
```

```python
# visualize
plt.imshow(binary_grid, cmap='gray')
plt.title('Binary Grid')
plt.axis('off')
plt.show()

print("Grid size:", binary_grid.shape)
```
